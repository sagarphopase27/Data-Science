```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
SP = pd.read_csv(r'C:\\Users\\Icon\\Documents\\Machine Learning Data\\Maharashtra RainFall.csv')
SP
```

```python
xpos = np.arange(len(SP['DISTRICT']))
xpos
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.xticks(xpos,SP['DISTRICT'],rotation=90, fontsize=10)
plt.title('Rainfall Map Maharashtra',fontsize=15,color='Red')
plt.xlabel('---DISTRICT---',fontsize=12, color='Red')
plt.ylabel('---Rainfall---',fontsize=15,color='Red')
plt.plot(xpos,SP.ANNUAL,label='Rainfall')
plt.grid()
plt.legend()
```
